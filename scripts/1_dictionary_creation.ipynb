## Take one law and create a dictionary based on article 2

```python
import pandas as pd
# Load the law

text = pd.read_excel('/Users/alexa/Projects/cdmx_kg/Mexico_City/LEY_AMBIENTAL_DE_LA_CIUDAD_DE_MEXICO.xlsx')['text'].iloc[3]
text
```

```python
import re

# Patrón regex para extraer las definiciones
# Busca: número romano + punto + espacio + término + : + definición
pattern = r'([CLXVI]+)\.\s+([^:]+?):\s+([^;]+(?:;[^;]*?(?=[CLXVI]+\.|$))*)'

# Encontrar todas las coincidencias
matches = re.findall(pattern, text)

# Crear el diccionario
definitions_dict = {}
definitions_list = []

for roman_num, term, definition in matches:
    # Limpiar espacios y caracteres extra
    term = term.strip()
    definition = definition.strip()
    
    # Agregar al diccionario
    definitions_dict[term] = definition
    
    # Agregar a la lista para el DataFrame
    definitions_list.append({
        'Numero': roman_num,
        'Termino': term,
        'Definicion': definition
    })

# Convertir a DataFrame
df_dictionary = pd.DataFrame(definitions_list)
print(f"Total de deficiones:{len(definitions_list)}")
df_dictionary.head()

df_dictionary.to_csv('/Users/alexa/Projects/cdmx_kg/data/dictionary_1.csv', index=False, encoding='utf-8-sig')
```
